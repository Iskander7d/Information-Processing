```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df = pd.read_csv('data70-30/grades_learn.csv')
df.head
```

```python
x = df.iloc[:,:8]
x.head()
```

```python
y=df.iloc[:,8:]
y.head()
```

```python
model=tree.DecisionTreeClassifier(criterion="entropy")
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
df_test = pd.read_csv('data70-30/grades_test.csv')
ll = df_test.values.tolist()
results = []
actual = []

for l in ll:
    results.append(model.predict([l[:8]]))
    actual.append(l[8:])

print(*actual)
print(*results)
```

```python
is_correct = 0
is_near = 0
for i in range(len(results)):
    if(results[i] == actual[i]):
        is_correct += 1
    if(abs(results[i] - actual[i]) <= 5):
        is_near += 1
    
print('accuracy: {}%'.format(100* is_correct/len(actual)))
print('close estimate: {}%'.format(100* is_near/len(actual)))
```

```python
plt.figure(figsize=(250,170))
tree.plot_tree(model, fontsize=100)
plt.show()
```
