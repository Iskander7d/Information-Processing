```python
import numpy as np
import pylab as pl
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
from sklearn.utils import shuffle
from matplotlib import pyplot as plt
import scipy.cluster.hierarchy as shc
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
```

```python
dataset = pd.read_csv('hiring.csv')
X = dataset.select_dtypes(exclude=['object']).values
#X = dataset.iloc[:, [2,3,4,5,6,11]].values
plt.figure(figsize=(10, 7))
plt.title("Hiring Dendograms")
plt.xlabel('respondents')
plt.ylabel('Euclidean distance')
dend = shc.dendrogram(shc.linkage(X, method='ward'))
```

```python
ac = AgglomerativeClustering(compute_full_tree=True, affinity='euclidean', linkage='ward')
y_ac = ac.fit_predict(X)
print(ac.n_clusters)
print(y_ac)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X[y_ac == 0, 0], X[y_ac == 0, 1], s = 100, c = 'blue', label = 'type 1')
plt.scatter(X[y_ac == 1, 0], X[y_ac == 1, 1], s = 100, c = 'orange', label = 'type 2')
plt.title('Clustering of customers')
plt.xlabel('Annual Income (k)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
